# Asia Cup Data Analysis


The Asian Cricket Council Asia Cup is a men's One Day International and Twenty20 International cricket tournament. It was established in 1983 when the Asian Cricket Council was founded as a measure to promote goodwill between Asian countries. It was originally scheduled to be held every two years. The Asia Cup is the only continental championship in cricket and the winning team becomes the champions of Asia. It alternates every 2 years between ODI and T20I formats.

The first Asia cup was held in 1984 in Sharjah in the United Arab Emirates where the council's offices were based (until 1995). India boycotted the 1986 tournament due to strained cricketing relations with Sri Lanka. Pakistan boycotted the 1990–91 tournament due to strained political relations with India and the 1993 tournament was cancelled for the same reason. The ACC announced that the tournament would be held biennially from 2009 onwards. The ICC has ruled that all games played in the Asia Cup have official ODI status.

After downsizing the Asian Cricket Council in 2015, it was announced by the ICC that Asia Cup events from 2016 would be played on a rotation basis between One Day International and Twenty20 International format, on the basis of the format of upcoming world events. As a result, the 2016 event was the first event played in the T20I format and functioned as a preparatory tournament ahead of the 2016 ICC World Twenty20.

## Exploratory Analysis and Visualization

**In this section we will be analyzing the data and try visualize it using graphs. We will be using the matplotlib and seaborn libraries to plot the graphs.**



Let's begin by importing`matplotlib.pyplot` and `seaborn`.

````python
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline

sns.set_style('darkgrid')
matplotlib.rcParams['font.size'] = 14
matplotlib.rcParams['figure.figsize'] = (9, 5)
matplotlib.rcParams['figure.facecolor'] = '#00000000'
````

 **To begin with let's see which team has won the most number of matches in the Asia Cup from 1984 to 2022**  

````python
win_counts = data[data['Result'] == 'Win']['Team'].value_counts()

plt.figure(figsize=(12, 6))
win_counts.plot(kind='bar')
plt.xlabel('Team')
plt.ylabel('Number of Wins')
plt.title('Number of Wins for Each Team')
plt.xticks(rotation=45)
plt.show()
````

Srilanka and India have won the most matches of the world cup with 39 wins by each team. Pakistan stands on third with 31 wins, Bangladesh on 4th with 9 wins and Afghanistan on 5th with only 5 wins. Hong Kong and UAE are yet to win a match in Asia Cup.

**Lets see how many times each team has scored 300+ runs in odi format and 200+ runs in t20i format**

````python
odi_data = data[(data['Format'] == 'ODI') & (data['Run Scored'] >= 300.0)]

odi_300_plus_counts = odi_data.groupby(['Team', 'Year']).size().unstack()

plt.figure(figsize=(12, 8))
odi_300_plus_counts.plot(kind='bar')
plt.xlabel('Team and Year')
plt.ylabel('Number of times 300.0+ scored (ODI)')
plt.title('Number of times each Team scored 300.0+ (ODI)')
plt.xticks(rotation=45)
plt.legend(title='Team', title_fontsize='12')
plt.show()
````

2008 was the year when most number of 300+ scores were made with India and Srilanka 4 times each and Pakistan and Bangladesh scored 300 or mores once in the year 2008.

````python
t20i_data = data[(data['Format'] == 'T20I') & (data['Run Scored'] >= 200.0)]

t20i_200_plus_counts = t20i_data.groupby(['Team', 'Year']).size().unstack()

plt.figure(figsize=(12, 8))
t20i_200_plus_counts.plot(kind='bar')
plt.xlabel('Team and Year')
plt.ylabel('Number of times 200.0+ scored (T20I)')
plt.title('Number of times each Team scored 200.0+ (T20I)')
plt.xticks(rotation=45)
plt.legend(title='Team', title_fontsize='12')
plt.show()
````

India is the only team to have scored 200+ in the T20I format of Asia Cup which came in 2022.

**Let's see how many times a team which has won the toss has also won the match**

````python
won_toss_won_match = data[(data['Toss'] == 'Win') & (data['Result'] == 'Win')]

team_counts = won_toss_won_match['Team'].value_counts()

plt.figure(figsize=(12, 8))
team_counts.plot(kind='bar')
plt.xlabel('Team')
plt.ylabel('Number of Matches')
plt.title('Number of Matches Won by Winning Toss')
plt.xticks(rotation=45)
plt.show()
````

Srilanka tops the chart here as well with 25 times when it won the toss and ended up winning the match as well.

**Let's analyze the batting strike rates of each team in Asia Cups**

````python
batting_strike_rate = data.groupby(['Year', 'Team'])['Avg Bat Strike Rate'].mean().unstack()

# Plot a bar chart for the average batting strike rate each year for each team
plt.figure(figsize=(20, 12))
batting_strike_rate.plot(kind='bar')
plt.xlabel('Year')
plt.ylabel('Average Batting Strike Rate')
plt.title('Average Batting Strike Rate Each Year Team Wise')
plt.xticks(rotation=45)
plt.legend(title='Team', title_fontsize='12')
plt.show()
````

As the years have progressed we can see an increase in the Average batting strike rate which clearly shows that teams now have started playing with more of an aggressive mind set. In the majority of years India had the higher Average Batting Strike Rate. The highest came in 2022 when Average Batting Strike for Indian Cricket Team was around 140.

**Let's see which player has won the Most Number of Man of the Match Awards for his team in the Asia Cups**

````python
unique_teams = data['Team'].unique()

for team in unique_teams:
    team_data = data[(data['Team'] == team) & (data['Result'] == 'Win')]

    if len(team_data) > 0:
        player_of_match_count = team_data['Player Of The Match'].value_counts()

        plt.figure(figsize=(10, 6))
        player_of_match_count.plot(kind='bar')
        plt.xlabel('Player Of The Match')
        plt.ylabel('Count')
        plt.title(f'Player Of The Match Count for {team}')
        plt.xticks(rotation=45)
        plt.show()
    else:
        print(f'{team} has not won any match')
````

Shoaib Malik and Sanath Jayasuriay (5 times each) from Pakistan and Srilanka respectively have won the most number of man of the matches for their teams. For India, it has been Virat Kholi who has won the most number of man of the match awards (6 times). Mahmudullah and Mushfiqur Rahim (twice each) have won the most number of man of the match awards for Bangladesh. For Afghanistan, 5 different players have won the man of the match award once.

## Asking and Answering Questions

Let's derive meaningful insights and provide well-informed responses by answering some questions



#### Q1: How does the run rate vary across different tournament years?

````python
average_run_rate = data.groupby('Year')['Run Rate'].mean()
````

````python
plt.figure(figsize=(12, 8))
average_run_rate.plot(kind='line', marker='o')
plt.xlabel('Year')
plt.ylabel('Average Run Rate')
plt.title('Average Run Rate Across Years')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
````

We can see a general increase in the Average Run Rate over the years with highest in 2022 with around 8 runs.

#### Q2: What is the distribution of the number of fours and sixes scored by each team?

````python
plt.figure(figsize=(12, 8))
sns.boxplot(data=data, x='Fours', y='Team', orient='h', palette='viridis')
plt.xlabel('Count')
plt.ylabel('Team')
plt.title('Distribution of Fours by Each Team')
plt.show()
````

Pakistan tops the chart here with most number of fours.

````python
plt.figure(figsize=(12, 8))
sns.boxplot(data=data, x='Sixes', y='Team', orient='h', palette='Blues')
plt.xlabel('Count')
plt.ylabel('Team')
plt.title('Distribution of Sixes by Each Team')
plt.show()
````

India has hit the most number of sixes.

#### Q3: What is the highest individual score by each team?

````python
highest_individual_score_by_team = data.groupby('Team')['Highest Score'].max()
````

````python
print(highest_individual_score_by_team)
````

#### Q4: What is the Average Wickets Taken by each Team?

````python
average_wickets_by_year_team = data.groupby(['Year', 'Team'])['Wicket Taken'].mean().unstack()

plt.figure(figsize=(12, 8))
sns.heatmap(average_wickets_by_year_team, cmap='YlGnBu', annot=True, fmt=".1f", linewidths=0.5)
plt.xlabel('Team')
plt.ylabel('Year')
plt.title('Average Wickets Taken by Each Team Each Year')
plt.show()
````

Across the years Pakistan and India have put up a really good show when it comes to bowling the team out. In the year 1997 and 2000 Pakistan averaged 10 wickets in each match. In 1984 India averaged 10 wickets in each match. In 2018 Srilanka averaged 10 wickets in each match.

#### Q5: How does the average batting strike rate vary between teams in different formats (ODI, T20I)?

````python
batting_strike_rate_data = data[['Team', 'Format', 'Avg Bat Strike Rate']]
plt.figure(figsize=(12, 8))
sns.boxplot(x='Team', y='Avg Bat Strike Rate', hue='Format', data=batting_strike_rate_data)
plt.xlabel('Team')
plt.ylabel('Average Batting Strike Rate')
plt.title('Average Batting Strike Rate Variation by Team and Format')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Format', loc='upper right')
plt.show()
````

**All teams play at a quick rate in T20I's. India seems to excel by some margin when it comes to strike rate in T20I's. However there doesnot seem much differrence in strike rate for Srilanka and Bangladesh.**
